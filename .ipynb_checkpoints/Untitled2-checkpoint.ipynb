```python
# import the usual resources
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = [1,2,3,4,5]
b = [0.5,1,4,5]
```

```python
# visualize the loss as the network trained
plt.plot(a,label='trianing_loss')
plt.plot(b,label='validation_loss')
plt.xlabel('1000\'s of batches')
plt.legend()
plt.ylabel('loss')
plt.ylim(0, 7) # consistent scale
plt.show()
```

```python
fig = plt.figure()
fig.savefig('config_1.png')
```
